```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
car2 = pd.read_csv('cars1.csv')
```

```python
car2.plot(kind='scatter', x='Selling_Price', y=
         'Present_Price')
```

```python
car2.columns
```

```python
car2.columns=car2.columns.str.strip()
```

```python
avg_price = car2.groupby('Fuel_Type')['Selling_Price'].mean()
avg_price.plot(kind='bar', title='Average Selling Price by Fuel Type')

plt.ylabel('Average Selling Price')
plt.xlabel('Fuel Type')
plt.show()
```
